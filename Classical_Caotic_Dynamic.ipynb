# Particle in a ring

### The Hamiltonian 
\begin{equation}
H = \frac{L^2}{2I} + k\cos \varphi \sum_{-\infty}^{\infty} \delta(t - n\tau)
\end{equation}

#### The motion equation
\begin{equation}
p_{n+1} = p_n + ksin \varphi
\end{equation}

\begin{equation}
x_{n+1} = x_n + p_{n+1}
\end{equation}

# Code

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Initial conditions
x_0 = np.random.random()%(2*np.pi)
p_0 = np.random.random()%(2*np.pi)
# time interval
t = 100

#caottic constant
k = 0.001

# solcing motions equations
result=[[x_0,p_0]]
for i in range(t):
    p = (p_0 + k*np.sin(x_0))%(2*np.pi)
    x = (p + x_0)%(2*np.pi)
    result.append([x,p])
    #reseting variables
    p_0 = p 
    x_0 = x
    
result = np.array(result)
```

```python
def trajectory(k_0):
    # Initial conditions
    x_0 = np.random.random()*(2*np.pi)
    p_0 = np.random.random()*(2*np.pi)
    # time interval
    t = 1000
    #caottic constant
    k = k_0
    
    # solcing motions equations
    result=[[x_0,p_0]]
    for i in range(t):
        p = (p_0 + k*np.sin(x_0))%(2*np.pi)
        x = (p + x_0)%(2*np.pi)
        result.append([x,p])
    #reseting variables
        p_0 = p 
        x_0 = x
       
    result = np.array(result)
    return result
```

```python
a = trajectory(.5)
```

```python
paths = 1000
a = np.array([trajectory(10) for i in range(paths)])
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
def trajectory(k_0):
    # Initial conditions
    x_0 = np.random.random()
    y_0 = np.random.random()
    z_0 = np.random.random()
    # time interval
    t = 1000
    #caottic constant
    k = k_0
    alf = 1
    
    
    # solcing motions equations
    result=[[x_0,y_0,z_0]]
    for i in range(t):
        x_bar = x_0
        y_bar = y_0*np.cos(alf) - z_0*np.sin(alf)
        z_bar = y_0*np.sin(alf) + z_0*np.cos(alf)
        
        x = x_0*np.cos(k*z_bar) - y_bar*np.sin(k*z_bar)
        y = x_0*np.sin(k*z_bar) + y_bar*np.cos(k*z_bar)
        z = z_bar
        #reseting variables
        x_0 = x
        y_0 = y
        z_0 = z
        result.append([x,y,z])
    result = np.array(result)
    return result
```

```python
a = trajectory(.1)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

   
   # Plot the values
ax.scatter(a[:,0], a[:,1], a[:,2], s=1)


plt.show()
```
